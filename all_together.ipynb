```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('reduced_stations_nonres.xlsx')
df.info()
```

```python
#df = df.drop(columns=['score_0_5_class'], axis=1)
```

```python
nonres_chosen = df.copy()
nonres_chosen = nonres_chosen.rename(columns={'lat': 'Latitude', 'lon': 'Longitude', 'final_score_0_25' : 'Score'})
nonres_chosen['Res/NonRes'] = 'N'
nonres_chosen.head()
```

```python
res_chosen = pd.read_csv('reduced_residential.csv')

res_chosen = res_chosen.drop(['PostcodeDistrict','CensusPopulationCount2022', 'SIMD2020_Quintile', 'Ward No.'], axis=1)
res_chosen = res_chosen.drop(['Unnamed: 0'], axis=1)
res_chosen['Res/NonRes'] = 'R'
res_chosen = res_chosen.rename(columns={'Postcode' : 'candidate_index'})
res_chosen.head(40)
```

```python
all_stations = pd.concat([nonres_chosen, res_chosen])
all_stations.info()
```

```python
all_stations.to_csv('all_stations_reduced_3.csv')
```
